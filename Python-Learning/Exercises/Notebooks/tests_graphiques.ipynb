```python
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np
```

### Définition de l'unité pixel

```python
px = 1 / plt.rcParams['figure.dpi']
```

### Définition de fonctions

```python
functions = [lambda n: n,
             lambda n: n*n,
             lambda n: np.sqrt(n),
             lambda n: pow(2, n)]
```

### Réprésentation de fonctions

```python
fig, axs = plt.subplots(figsize=(1000*px, 300*px))
x        = np.linspace(0, 7, 80)

for f in functions:
    axs.plot(x, f(x))

plt.grid(linestyle='--')
plt.show()
```

### Réprésentation des données

```python
fig, axs = plt.subplots(2, 1, figsize=(1200*px, 600*px))

X        = np.arange(0, 200, 1)
Y        = np.random.randint(100, size=200)

xticks_0 = np.arange(0, 201, 10)
yticks_0 = np.arange(0, 101, 10)

axs[0].set_xticks(xticks_0)
axs[0].set_yticks(yticks_0)
axs[0].scatter(X, Y, s=1, c='r')

xticks_1 = np.arange(0, 101, 5)
yticks_1 = np.arange(0, 31, 5)

axs[1].set_xticks(xticks_1)
axs[1].set_yticks(yticks_1)
axs[1].hist(Y, bins=np.arange(0, 101, 10), ec='black')

plt.show()
```

### Déplacement de balle

```python
%matplotlib ipympl

fig, axs = plt.subplots(figsize=(1000*px, 300*px))
ball     = plt.Circle((100, 150), 15, fc='b')

axs.set_xlim(0, 1000)
axs.set_ylim(0, 300)
axs.add_patch(ball)

dx      = 10

def update(i):
    global dx
    x, y        = ball.center
    dx          = -dx if x < 100 or x > 900 else dx
    x           += dx
    ball.center = (x, y)

loop    = animation.FuncAnimation(fig, update, interval=33, blit=True)

plt.show()
```
